```python
import pandas as pd

df = pd.read_csv("A:/employee_pfm.csv")
empqtr = df.groupby(['quarter_Quarter1', 'quarter_Quarter2', 'quarter_Quarter3', 'quarter_Quarter4', 'quarter_Quarter5']).size()
grpqrt=pd.DataFrame(empqtr)
grpqrt
```

```python
#employee counts by day
day_counts = df[['day_Monday', 'day_Tuesday', 'day_Wednesday', 'day_Thursday', 'day_Saturday', 'day_Sunday']].sum()
print("Employee counts by day:")
dc=pd.DataFrame(day_counts)
dc
```

```python
# Filter by Departments
finishing_dept = df[df['department_finishing'] == 1]
sweing_dept = df[df['department_sweing'] == 1]

print(f"Finishing department employees count: {len(finishing_dept)}")
print(f"Sweing department employees count: {len(sweing_dept)}")
```

```python
df['Total_Days_Worked'] = df[['day_Monday', 'day_Tuesday', 'day_Wednesday', 'day_Thursday', 'day_Saturday', 'day_Sunday']].sum(axis=1)
da=df.head()
frame=pd.DataFrame(da)
frame
```

```python
deptsmry = df.groupby(['department_finishing', 'department_sweing'])['Total_Days_Worked'].sum()
print("Total Days Worked per Department:")
dw=pd.DataFrame(deptsmry)
dw
```
